```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price (INR)'], kde=True, color='skyblue')
plt.title('Distribution of Bike Prices')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.show()
# Distribution of Bike Prices
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='State', y='Avg Daily Distance (km)', palette='Set2')
plt.xticks(rotation=45)
plt.title('Average Daily Distance by State')
plt.show()
# How does average daily distance vary by state?
```

```python
df['Brand'].value_counts().plot(kind='bar', color='coral', figsize=(10, 5))
plt.title('Most Popular Bike Brands')
plt.ylabel('Count')
plt.xlabel('Brand')
plt.xticks(rotation=45)
plt.show()
# Most Popular Bike Brands
```

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Depreciation (%)'] = ((df['Price (INR)'] - df['Resale Price (INR)']) / df['Price (INR)']) * 100
sns.histplot(df['Depreciation (%)'], bins=20, kde=True, color='purple')
plt.title('Distribution of Depreciation (%)')
plt.show()
# How does resale price compared to original price?
```

```python
sns.countplot(data=df, x='Fuel Type', palette='Set1')
plt.title('Fuel Type Distribution')
plt.show()
# Bike counts by fuel type
```

```python
sns.boxplot(data=df, x='Seller Type', y='Resale Price (INR)', palette='Set3')
plt.title("Resale Price by Seller Type")
plt.show()
```

```python
df.groupby('Fuel Type')['Mileage (km/l)'].mean().plot(kind='bar', color='teal')
plt.ylabel("Average Mileage (km/l)")
plt.title("Average Mileage by Fuel Type")
plt.show()
```

```python
sns.violinplot(data=df, x='City Tier', y='Price (INR)', palette='pastel')
plt.title("Bike Prices Across City Tiers")
plt.show()
```

```python
df.groupby('Year of Manufacture')['Price (INR)'].mean().plot(marker='o', color='green')
plt.title("Average Price by Year of Manufacture")
plt.ylabel("Avg Price (INR)")
plt.grid(True)
plt.show()
```

```python
sns.countplot(data=df, x='Owner Type', order=df['Owner Type'].value_counts().index, palette='cubehelix')
plt.title("Bike Count by Owner Type")
plt.show()
```

```python
sns.scatterplot(data=df, x='Mileage (km/l)', y='Resale Price (INR)', hue='Fuel Type', palette='pastel')
plt.title("Mileage vs Resale Price")
plt.show()
```

```python
sns.boxplot(data=df, x='Insurance Status', y='Resale Price (INR)', palette='cool')
plt.title("Resale Value by Insurance Status")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Brand', y='Engine Capacity (cc)', palette='Set1')
plt.title("Engine Capacity Across Brands")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Registration Year'].value_counts().sort_index().plot(kind='bar', color='slateblue')
plt.title("Bike Registrations Over the Years")
plt.xlabel("Registration Year")
plt.ylabel("Count")
plt.show()
```

```python
sample_df = df.sample(30)
plt.figure(figsize=(12,6))
plt.plot(sample_df['Model'], sample_df['Price (INR)'], label='Original Price', color='orange')
plt.plot(sample_df['Model'], sample_df['Resale Price (INR)'], label='Resale Price', color='blue')
plt.xticks(rotation=90)
plt.legend()
plt.title("Original vs Resale Prices for Sample Bikes")
plt.show()
```

```python
top_mileage = df[['Model', 'Mileage (km/l)']].drop_duplicates(subset='Model')
top_mileage = top_mileage.sort_values(by='Mileage (km/l)', ascending=False).head(10)

sns.barplot(x='Mileage (km/l)', y='Model', data=top_mileage, palette='Greens_r')
plt.title('Top 10 High Mileage Bike Models')
plt.xlabel('Mileage (km/l)')
plt.ylabel('Bike Model')
plt.tight_layout()
plt.show()

```
